I do it for any number now it has two parts, integer and fractional part.

```python
def decimal_to_binary(num):
    int_part = int(num) # integer part
    frac_part = num - int_part  # fractional 
    int_binary = ""
    if int_part == 0:
        int_binary = "0"
    else:
        while int_part > 0:
            remainder = int_part % 2
            int_binary = str(remainder) + int_binary  # Add remainder at the beginning
            int_part = int_part // 2

    # Convert fractional part to binary
    frac_binary = ""
    p = 14
    while frac_part > 0 and len(frac_binary) < p:
        frac_part*= 2
        bit= int(frac_part)
        frac_binary= frac_binary + str(bit)
        frac_part= frac_part - bit 

    # Combine integer and fractional binary parts
    if frac_binary:
        return f"{int_binary}.{frac_binary}"
    else:
        return int_binary

decimal_number = 9 # Replace with any decimal number
binary = decimal_to_binary(decimal_number)
print(f"Binary representation of {decimal_number} is {binary}")
```

(4.2) to convert a decimal type to s, e, f.

```python
def decimal_to_float(number):
    # Handle special cases
    if abs(number) > 1e40 :
        return (0 if number > 0 else 1), 0xFF, 0  # Infinity: s=sign, e=all 1s, f=0
    if number == 0:
        return 0, 0, 0  # Zero: s=0, e=0, f=0
    
    s = 0 if number > 0 else 1
    number = abs(number)
    
    e = 0
    if number >= 1:
        while number >= 2:
            number /= 2
            e += 1
    elif number < 1:
        while number < 1:
            number *= 2
            e -= 1

    # Adjust exponent to fit the bias (127 for single-precision)
    e = e + 127
    print(number)
    number -= 1 #subtracting 1 we are left with just the fractional part.
    f= ""
    for _ in range(23):
        number *= 2
        bit = int(number)
        f = f + str(bit)
        number = number - bit
    return s, e, f

# Example usage
number = 3.14
s, e, f = decimal_to_float(number)
print(f"Number: {number}")
print(f"Sign (s): {s}, Exponent (e): {e}, Fraction/Mantissa (f): {f}")
1e2
```

(1) the code for factorial of any number-


```python
def factorial(n):
    f=1
    while n>0:
        f=f*n
        n=n-1
    return f
factorial(5)
```
